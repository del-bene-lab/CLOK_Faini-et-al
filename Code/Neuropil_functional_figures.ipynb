## Analysis for Dins and Dins

SINs and DINs are two neuronal populations located within the tectal neuropil. SINs reside closer to the skin surface, whereas DINs are positioned deeper within the neuropil. Here, we co-expressed H2B-GCaMP6s and H2B-HaloTag in neurons by crossing elavl3:H2B-GCaMP6s fish with xla-tubb2:hsp70-ubc-H2B-HaloTag fish. We then performed double labeling with JF552 and JF669 at 2 and 5 dpf. Neuronal activity was recorded at 6 dpf using two-photon calcium imaging during a series of visual stimuli.

This code performs essentially the same analyses as tectum_functional_analysis.ipynb, but on a different dataset. In this dataset, we have the position, red fluorescence, and far-red fluorescence of SINs and DINs, along with their activity traces recorded during the battery of visual stimulations.

```python
def spearman(dfof, ratio, stat = True):
    if stat:
        return sc.spearmanr(dfof, ratio)[0]
    else:
        return sc.spearmanr(dfof, ratio)[1]

def critical_r(n, alpha=0.05, tail='two-sided'):

    """
    Studentâ€™s t-distribution is used to calculate the critical value for determining the significance threshold.
 Note that this function is used only to draw the threshold line on the graph; SciPy calculates the threshold on its own to determine significance
    """
    df = n - 2
    if tail == 'two-sided':
        t_crit = t.ppf(1 - alpha/2, df)
    elif tail == 'one-sided':
        t_crit = t.ppf(1 - alpha, df)
    else:
        raise ValueError("tail must be 'one-sided' or 'two-sided'")
    
    r_crit = t_crit / np.sqrt(t_crit**2 + df)
    return r_crit
```

```python
def colored_line(x, y, c, ax, **lc_kwargs):
    """
    From matplotlib's documentation
    """
    
    if "array" in lc_kwargs:
        warnings.warn('The provided "array" keyword argument will be overridden')

    default_kwargs = {"capstyle": "butt"}
    default_kwargs.update(lc_kwargs)

    x = np.asarray(x)
    y = np.asarray(y)
    x_midpts = np.hstack((x[0], 0.5 * (x[1:] + x[:-1]), x[-1]))
    y_midpts = np.hstack((y[0], 0.5 * (y[1:] + y[:-1]), y[-1]))

    coord_start = np.column_stack((x_midpts[:-1], y_midpts[:-1]))[:, np.newaxis, :]
    coord_mid = np.column_stack((x, y))[:, np.newaxis, :]
    coord_end = np.column_stack((x_midpts[1:], y_midpts[1:]))[:, np.newaxis, :]
    segments = np.concatenate((coord_start, coord_mid, coord_end), axis=1)

    lc = LineCollection(segments, **default_kwargs)
    lc.set_array(c)  # set the colors of each segment

    return ax.add_collection(lc)
```
